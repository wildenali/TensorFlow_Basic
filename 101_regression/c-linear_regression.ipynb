```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
```

```python
x, y = datasets.make_regression(n_features=1, n_samples=300)
```

```python
y = y.reshape(-1,1)
```

```python
x.shape, y.shape
```

```python
n_sample, n_features = x.shape
model = tf.keras.models.Sequential()
model.add(tf.keras.layers.Dense(1))
optimizer = tf.keras.optimizers.SGD(learning_rate=0.1)
```

```python
model.compile(optimizer=optimizer, loss="mean_absolute_error") # untuk compalisasi bentuk model
```

```python
model.fit(x, y, verbose=1, epochs=50)
```

```python
y_pred = model.predict(x)
plt.plot(x, y, 'ro')
plt.plot(x, y_pred, 'bo')
plt.show()
```

```python
x, y = datasets.make_regression(n_features=5, n_samples=300)
y = y.reshape(-1,1)
x.shape, y.shape

n_sample, n_features = x.shape
model = tf.keras.models.Sequential()
model.add(tf.keras.layers.Dense(5))
model.add(tf.keras.layers.Dense(3))
model.add(tf.keras.layers.Dense(1))
optimizer = tf.keras.optimizers.SGD(learning_rate=0.1)
model.compile(optimizer=optimizer, loss="mean_absolute_error") # untuk compalisasi bentuk model
model.fit(x, y, verbose=1, epochs=70)
y_pred = model.predict(x)
plt.plot(x, y, 'ro')
plt.plot(x, y_pred, 'bo')
plt.show()
```
