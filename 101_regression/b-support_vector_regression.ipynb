```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVR
from sklearn.metrics import mean_squared_error as mse
```

```python
data = datasets.load_diabetes() # https://scikit-learn.org/stable/datasets/toy_dataset.html#diabetes-dataset
```

```python
data.feature_names
```

```python
data.feature_names, data.DESCR
```

```python
X = data.data
Y = data.target
```

```python
X.shape, Y.shape
```

```python
X[0]
```

```python
X[0], Y[0]
```

```python
# Normalize Data
sc = StandardScaler()
X = sc.fit_transform(X)
```

<img src="svr_kernel.png">

## Coba pakai kernel="linear"

```python
svreg = SVR(kernel="linear")
svreg.fit(X, Y)
```

```python
ypred = svreg.predict(X)
```

```python
plt.figure(figsize=(14,9))
plt.plot(Y, color='red')
plt.plot(ypred, color='green')
plt.show()
```

```python
hasil_mse = mse(Y, ypred)
```

```python
print(hasil_mse)
```

## Coba pakai kernel="rbf"

```python
svreg = SVR(kernel="rbf", gamma=0.3)
svreg.fit(X, Y)
ypred = svreg.predict(X)
plt.figure(figsize=(14,9))
plt.plot(Y, color='red')
plt.plot(ypred, color='green')
plt.show()
```

```python
hasil_mse = mse(Y, ypred)
print(hasil_mse)    # hasil prediksi dengan nilai aslinya sangat jauh
```

gamma adalah hyperparameter
nilai C adalah variabel untuk reguler decision, lebih secara general mengolah data
